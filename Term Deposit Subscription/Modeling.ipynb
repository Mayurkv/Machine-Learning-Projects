# Cross Validation Strategy

```python
def cross_val(train, test, features, clf):
    N_splits = 7
    
    preds_df = pd.DataFrame()
    
    oofs = np.zeros(len(train))
    preds = np.zeros(len(test))
    
    target_col = train[target]
    
    folds = StratifiedKFold(n_splits = N_splits, shuffle = True, random_state = 5)
    stratified_target = pd.qcut(target_col, 10, labels=False, duplicates='drop')
    
    for index, (trn_idx, val_idx) in enumerate(folds.split(train,stratified_target)):
        print()
        print(f'****************Fold {index+1}****************')
        
        ## Training Data
        X_trn, y_trn = train[features].iloc[trn_idx], train[target].iloc[trn_idx]
        
        ## Validation data
        X_val, y_val = train[features].iloc[val_idx], train[target].iloc[val_idx]
        
        ## Test Data
        X_test = test[features]
        
        # Model Training
        _ = clf.fit(X_trn, y_trn)
        
        # Predicting
        val_preds = clf.predict_proba(X_val)[:, 1]
        test_preds = clf.predict_proba(X_test)[:, 1]
        
        oofs[val_idx] = val_preds
        #preds_df[str(index+1)] = test_preds
        preds += test_preds/N_splits
        
        f1_score_ = f1_score(y_val, val_preds.round())
        print()
        print(f'F1-Score is {f1_score_}')
    
    total_score = f1_score(target_col, oofs.round())
    print()
    print(f'F1-Score is {total_score}')
    ## Aggregating Test Predictions
    #preds_df['Mode'] = preds_df.apply(lambda x: x.mode()[0], axis = 1)
    #preds = preds_df['Mode']
    return oofs, preds
```

```python
lgb_oofs, lgb_preds = cross_val(train_proc, test_proc, features, LGBMClassifier(random_state=5))
```

```python
sample[target] = lgb_preds
sample.to_csv(r"D:\Data Science\term deposit subscription\lgb cross val.csv", index=False)
```

```python
def feat_engg(df, x, y):
    df[f'{y}_mean'] = df.groupby(x)[y].transform('mean')
    df[f'{y}_median'] = df.groupby(x)[y].transform('median')
    df[f'{y}_min'] = df.groupby(x)[y].transform('min')
    df[f'{y}_max'] = df.groupby(x)[y].transform('max')
    df[f'{y}_std'] = df.groupby(x)[y].transform('std')
    
    return df
```

```python
def get_optimal_thresh(target, oofs):
    thresholds = np.arange(0,100)/100
    thresh_scores = []
    for thresh in thresholds:
        oofs_rounded = (oofs>thresh)*1
        thresh_score = f1_score(target, oofs_rounded)
        thresh_scores.append(thresh_score)
    
    all_thresh_scores = pd.Series(index=thresholds, data=thresh_scores)
    return all_thresh_scores.sort_values(ascending=False).index.values[0]
    #return all_thresh_scores
```

```python
thresh = get_optimal_thresh(train[target], lgb_oofs_3)
```
